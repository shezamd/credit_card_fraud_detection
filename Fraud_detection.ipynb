# Preprocess Data

```python
df.head(3)
```

```python
# Select features and labels
# Drop columns with correlation 
X = df.drop(columns=['Time', 'Class', 'V13', 'V15', 'V22', 'V23', 'V25', 'V26'])
# X = df.drop(columns=['Time', 'Class'])
y = df['Class']

# split data 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test  = train_test_split(X, y, test_size=0.2)


print('Overall Data:', len(X))
print('Train Data:  ', len(X_train))
print('Test Data:   ', len(X_test))
```

```python
X.head(2)
```

# Catboost

```python
# pip install catboost
```

```python
from catboost import CatBoostClassifier

model = CatBoostClassifier(iterations=200, eval_metric='AUC', verbose=False)

model.fit(X_train, y_train, plot=True)

predictions = model.predict(X_test)
```

# Model Evalution Metric

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score, f1_score

# Train
train_predictions = model.predict(X_train)

train_accuracy  = round(accuracy_score(y_train, train_predictions)*100, 2)
train_roc_auc   = round(roc_auc_score(y_train, train_predictions)*100, 2)
train_precision = round(precision_score(y_train, train_predictions)*100, 2)
train_recall    = round(recall_score(y_train, train_predictions)*100, 2)
train_F1        = round(f1_score(y_train, train_predictions)*100, 2)

print('Train Accuracy  :', train_accuracy)
print('Train ROC AUC   :', train_roc_auc)
print('Train Precision :', train_precision)
print('Train Recall    :', train_recall)
print('Train F1        :', train_F1)
print()


# Test
test_accuracy = round(accuracy_score(y_test, predictions)*100, 2)
roc_auc       = round(roc_auc_score(y_test, predictions)*100, 2)
precision     = round(precision_score(y_test, predictions)*100, 2)
recall        = round(recall_score(y_test, predictions)*100, 2)
F1            = round(f1_score(y_test, predictions)*100, 2)

print('Test Accuracy   :', accuracy)
print('Test ROC AUC    :', roc_auc)
print('Test Precision  :', precision)
print('Test Recall     :', recall)
print('Test F1         :', F1)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

print('Training Confusion Matrix')
cm = confusion_matrix(y_train, train_predictions, labels=model.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot()
plt.grid()
plt.show()
```

```python
print('Testing Confusion Matrix')
cm = confusion_matrix(y_test, predictions, labels=model.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot()
plt.grid()
plt.show()
```

```python
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.sort_values(ascending=False).plot(kind='bar', 
                                                   figsize=(15, 5), 
                                                   title='Feature Importance', 
                                                   xlabel='Feature', 
                                                   ylabel='Importance')
```
